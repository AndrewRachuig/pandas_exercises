### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
tips
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
tips
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips.total_bill)
sns.despine()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip)
sns.despine()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.scatterplot(x = tips.total_bill, y = tips.tip, data = tips, hue = 'size')
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(12,10))
sns.boxplot(data = tips, y = 'total_bill', x = 'day')
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(data = tips, y='day', x= 'tip', hue = 'sex')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data = tips, y= 'total_bill', x = 'day', hue = 'time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(12,8))
plt.title('Lunch v Dinner tips')
plt.subplot(121)
sns.histplot(data = tips[tips.time == 'Dinner'], x = 'tip')
plt.ylim(0,40)
plt.xlim(0,11)
sns.despine()

plt.subplot(122)
sns.histplot(data = tips[tips.time == 'Lunch'], x = 'tip')
plt.ylim(0,40)
plt.xlim(0,11)
sns.despine();
```

```python
# This is the provided solution, mine is the cell above. I prefer mine for being more consistent and fair with the data.
g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(12,8))
plt.title('Lunch v Dinner tips')
plt.subplot(121)
sns.scatterplot(data = tips[tips.sex == 'Male'], x = 'total_bill', y ='tip', hue = 'smoker', s = 70, alpha= .7)
plt.ylim(0,11)
plt.xlim(0,50)
sns.despine()

plt.subplot(122)
sns.scatterplot(data = tips[tips.sex == 'Female'], x = 'total_bill', y ='tip', hue = 'smoker', s = 70, alpha= .7)
plt.ylim(0,11)
plt.xlim(0,50)
sns.despine();
```
